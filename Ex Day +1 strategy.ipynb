```python
def FigReturns(stocks):
    outfile = open("returns.csv",'w',newline='')
    returns = csv.writer(outfile)
    for stock in stocks['Symbol']:
        #stock = str(stock)
        print (stock)
        a = requests.get("https://api.iextrading.com/1.0/stock/" + stock + "/chart/5y")
        b = requests.get("https://api.iextrading.com/1.0/stock/" + stock + "/dividends/5y")
        prices = pd.read_json(a.text)
        divs = pd.read_json(b.text)
        divs.exDate = pd.to_datetime(divs.exDate)
        prices['NextDay'] = prices['date']- pd.Timedelta(days=1)
        divprice = pd.merge(divs,prices,left_on= "exDate",right_on= ("NextDay"))
        meanreturn = divprice.changePercent.mean()
        
        
        returns.writerow((stock,meanreturn))
        
        
    #returns.to_csv("returns.csv")
    
    outfile.close()
        
        
        
        
```
